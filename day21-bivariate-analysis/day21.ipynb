```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

## 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
```

## 2. Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
```

## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

## 4. Distplot (Numerical - Categorical)

```python
sns.displot(titanic[titanic['Survived']==0]['Age'])
sns.displot(titanic[titanic['Survived']==1]['Age'])
```

## 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Embarked').mean()['Survived']*100
```

## 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

## 7. Pairplot

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

## 8. Lineplot (Numerical - Numerical)

```python
flights.head()
```

```python
new = flights.groupby('year').sum(numeric_only=True).reset_index()
new
```

```python
sns.lineplot(x=new['year'],y=new['passengers'])
```
